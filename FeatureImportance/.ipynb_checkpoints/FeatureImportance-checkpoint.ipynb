## Data

In our synthetic dataset lets choose 1,000 examples each with 10 input variables, five of which are redundant and five of which are important to the outcome.

```python
# define dataset size and features
sample_size = 1000
n_num_features = 10
feature_name = 'col_{}'
target_name = 'target'
state = 1
redundany = 5
informative_features = 5
    
def create_date(purpose=''):
    if purpose == 'classification':
        print('generating classification data')
        X, y = make_classification(n_samples=sample_size, n_features=n_num_features, 
                                   n_informative=informative_features, n_redundant=redundany, 
                                   random_state=state)
    else:
        print('generating regression data')
        X, y = make_regression(n_samples=sample_size, n_features=n_num_features, 
                               n_informative=informative_features, random_state=state)
    return X, y

def get_df(X,y):
    # create a pandas df for future need or just for practice.
    df1 = pd.DataFrame(X,columns=[feature_name.format(i) for i in range(n_num_features)])
    df2 = pd.DataFrame(pd.Series(y, name=target_name), columns=[target_name])
    df = df1.merge(df2, left_index=True, right_index=True)
    return df1, df2, df
    
```

```python
X, y = create_date('classification')
print(X.shape, y.shape)

df1, df2, df = get_df(X,y)
display(df.head())
print(df.shape)
```

```python
#check if data is balanced or not
classes = np.unique(y)
class_weights=compute_class_weight('balanced', classes, y)
print('class balance = ', class_weights, 'for classes', classes)
```

## Utility Functions

```python
def plotFeatureImp(model, ensemble, mod_type='', scoring_critera=''):
    # get importance and plot it
    
    if ensemble:        
        if mod_type == 'kmean':
            # perform permutation importance against a scoring criteria and get mean importance
            results = permutation_importance(model, X, y, scoring='neg_mean_squared_error')
            importance = results.importances_mean
        elif mod_type == 'fs_with_imp':
            importance = model.estimator_.feature_importances_
        else:
            importance = model.feature_importances_
            #std = np.std([model.feature_importances_ for tree in model.estimators_], axis=0)
    else:
        if mod_type == 'logReg':
            importance = model.coef_[0]
        else: 
            importance = model.coef_ 
        
    #std = np.std([importance for tree in model.estimators_], axis=0)  
    indices = np.argsort(importance)[::-1]
    
    
    # Create a dataframe to map features to their importances
    features = pd.DataFrame()
    features['feature'] = df1.columns
    features['importance'] = importance
    
    # Print out important features
    df_fet = features[ features.importance > np.mean(features.importance) ].sort_values(by='importance', ascending=False)
    display(df_fet)
        
    # Plot the feature importances of the forest
    plt.figure()
    plt.title("Feature importances")
    plt.bar(range(X.shape[1]), importance[indices], color="b", align="center")
    plt.xticks(range(X.shape[1]), indices)
    plt.xlim([-1, X.shape[1]])
    plt.show()

    #for i,v in enumerate(importance):
    #    print('Feature: %0d, Score: %.2f' % (i,v))
    #plt.bar([x for x in range(len(importance))], importance)
    #plt.show()
```

## Permutation Feature Importance

It is a technique for calculating relative importance scores independent of the model.

First, a model is fit on the dataset, such as a model that does not support native feature importance scores, e.g., `k-nearest neighbors`. Then the model is used to make predictions on a dataset, although the values of a feature (column) in the dataset are scrambled. This is repeated for each feature in the dataset. Then this whole process is repeated 3, 5, 10 or more times. The result is a **mean importance score** for each input feature (and distribution of scores given the repeats).

This approach can be used for regression or classification and requires that a performance metric be chosen as the basis of the importance score, such as the **mean squared error for regression** and **accuracy for classification**.

Permutation feature selection can be used via the `permutation_importance()` function that takes a fit model, a dataset (train or test dataset is fine), and a scoring function.

## Feature Selection with Importance

Recall, our synthetic dataset has 1,000 examples each with 10 input variables, five of which are redundant and five of which are important to the outcome. We can use feature importance scores to help select the five variables that are relevant and only use them as inputs to a predictive model.

First, we can split the training dataset into train and test sets and train a model on the training dataset, make predictions on the test set and evaluate the result using classification accuracy. We will use a logistic regression model as the predictive model. This provides a baseline for comparison when we remove some features using feature importance scores.

```python
# split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
# fit the model
model = LogisticRegression(solver='liblinear')
model.fit(X_train, y_train)
# evaluate the model predictions
yhat = model.predict(X_test)
accuracy = accuracy_score(y_test, yhat)
print('Accuracy: %.2f' % (accuracy*100))
```

Given that we created the dataset, we would expect better or the same results with half the number of input variables. 

We could use any of the feature importance scores explored above, but in this case we will use the feature importance scores provided by random forest. We can use the `SelectFromModel` class to define both the model we wish to calculate importance scores, `RandomForestClassifier` in this case, and the number of features to select, `5` in this case.

The bottom result shows, that same level of accuracy can be achieved by the first 5 important features. This is somewhat analogous to SVM as dimensionality reduction case.

```python
from sklearn.feature_selection import SelectFromModel

# feature selection
def select_features(X_train, y_train, X_test):
    # configure to select a subset of features
    fs = SelectFromModel(RandomForestClassifier(n_estimators=1000), max_features=5)
    # learn relationship from training data
    fs.fit(X_train, y_train)
    # transform train input data
    X_train_fs = fs.transform(X_train)
    # transform test input data
    X_test_fs = fs.transform(X_test)
    return X_train_fs, X_test_fs, fs
 
# split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
# feature selection
X_train_fs, X_test_fs, fs = select_features(X_train, y_train, X_test)

print('train data', X_train_fs.shape)

print('test data', X_test_fs.shape)

print(fs)
plotFeatureImp(model=fs, ensemble=1, mod_type='fs_with_imp', scoring_critera='')

# fit the model
model = LogisticRegression(solver='liblinear')
model.fit(X_train_fs, y_train)
# evaluate the model
yhat = model.predict(X_test_fs)
# evaluate predictions
accuracy = accuracy_score(y_test, yhat)
print('Accuracy: %.2f' % (accuracy*100))
```

# Shapely Values

For comparison, lets reproduce the feature importance using RFC.

```python
X=df1

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)
rfc = RandomForestClassifier(n_estimators=1000)
rfc.fit(X_train, y_train)
plotFeatureImp(model=rfc, ensemble=1, mod_type='', scoring_critera='')
```

## eli5

Lets compare the results using the class `eli5`, which produces the exact same table as what we generated by hand. However, there is a slight difference between the weights or feature importance.

```python
import eli5
from eli5.sklearn import PermutationImportance

perm = PermutationImportance(rfc, random_state=1).fit(X_test, y_test)
cols = X_test.columns.tolist()
eli5.show_weights(perm, feature_names = cols)
```

### Check model accuracy and probability for classifying class=1

```python
yhat = rfc.predict(X_test)
print('The accuracy of the Random Forests model is : %.2f\t'% (accuracy_score(yhat,y_test)*100))
#print probability of getting class 1
probs = rfc.predict_proba(X_test)[:,1]#.reshape(1, -1)
# print probability of getting class 1 for the first element in test data
print(probs[:10])
```

### Plot to show feature importance of one element of the Test Data.

```python
import shap 
explainer = shap.TreeExplainer(rfc)
print('input features:\n', X_test.iloc[1], '\n\n true label:', y_test[1], '\n')
# Calculate Shap values for any instance in the test data
choosen_instance = X_test.iloc[1] # entry 1
shap_values = explainer.shap_values(choosen_instance)

# for class 0
print('base value: %.2f'% explainer.expected_value[0]) 
print('shap_values[0]: ', shap_values[0])
      
# for class 1     
print('\nbase value: %.2f'% explainer.expected_value[1]) 
print('shap_values[1]: ', shap_values[1])

shap.initjs()
shap.force_plot(explainer.expected_value[1], shap_values[1], choosen_instance)
```

The function `KernelExplainer()` below performs a local regression by taking the prediction method `rfc.predict` and the data that you want to perform the SHAP values.  If the prediction method `rfc.predict_proba` is used, then that will plot the probability.

```python
rf_shap_values = shap.KernelExplainer(rfc.predict, X_test.iloc[:10])
k_shap_values = rf_shap_values.shap_values(X_test.iloc[:10])
shap.summary_plot(k_shap_values, X_test.iloc[:10])
```

```python
rf_shap_values = shap.KernelExplainer(rfc.predict_proba, X_test.iloc[:10])
k_shap_values = rf_shap_values.shap_values(X_test.iloc[:10])
shap.summary_plot(k_shap_values, X_test.iloc[:10])
```

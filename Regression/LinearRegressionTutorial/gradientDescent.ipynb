# Gradient Descent
Gradient descent is a first-order iterative optimization algorithm for finding the minimum of a function. To find a local minimum of a function using gradient descent, one takes steps proportional to the negative of the gradient (or approximate gradient) of the function at the current point.

Example: For a function $f(x)=x^{4}-3x^{3}+2$, with derivative $f'(x)=4x^{3}-9x^{2}$, the gradient descent algorithm will solve for $f'(x)=0$ which solutions shows the function has a plateau point at 0 and a global minimum at $x={\tfrac {9}{4}}=2.25$.

```python
import matplotlib.pyplot as plt

next_x = 6  # We start the search at x=6
gamma = 0.01  # Step size multiplier
precision = 0.00001  # Desired precision of result
max_iters = 10000  # Maximum number of iterations

# Derivative function
df = lambda x: 4 * x**3 - 9 * x**2

dataX = []
dataY = []
steps = []
mins = []
for i in range(max_iters):
    #print (i)
    dataX.append(i)
    dataY.append(df(i))
    current_x = next_x
    next_x = current_x - gamma * df(current_x) # steps are prop. to neg. of the gradient
    mins.append(next_x)
    step = next_x - current_x
    steps.append(step) 
    if abs(step) <= precision:
        break
      
fig,(ax0,ax1, ax2) = plt.subplots(1,3, figsize=(15,5))
ax0.scatter(dataX, dataY,label='derivative')
ax1.plot(dataX, mins, color='red', label='minimum',linewidth=1.0)
ax2.plot(dataX[1:], steps[1:], color='blue', label='steps',linewidth=1.0)
ax0.legend()
ax1.legend()
ax2.legend()
fig.suptitle('Gradient Descent')
plt.show()
print("Number of steps", len(steps))
print("Minimum at", next_x)
```
